## Import das bibliotecas

```
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report
from sklearn import tree
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.neural_network import MLPClassifier
```

## Conexão com o Google Drive e import do dataset

```
from google.colab import drive

# conexão do colab com o google drive
drive.mount('/content/drive')

# criando um datafram com a primeira aba do dataset
sonar = pd.read_excel('/content/drive/My Drive/sonar.xlsx', sheet_name=0)
```

```
sonar.head()
```

## Tratamento no dataset para aplicar os modelos

```
# passando todas as linhas na variavel X usando iloc para ler todas as informações
X = sonar.iloc[:, 0:(sonar.shape[1] - 1)]

# aplicando label enconder para converter variaveis categoricas em númericas e treinando o modelo em todas as features(colunas)
le = LabelEncoder()
y = le.fit_transform(sonar.iloc[:, (sonar.shape[1] - 1)])

# gravando num objetvo os nomes das classes
class_names = le.classes_
```

```
# dividindo as bases em bases de treino e test com 10% para test e 90% para treino
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 0, test_size = 0.1)
```

## Decision tree

```
# chamando o metodo decisiontreeclassifier
decision_tree = tree.DecisionTreeClassifier()

# treinando a base de treino
decision_tree = decision_tree.fit(X_train, y_train)

# prevendo a base de test
y_pred = decision_tree.predict(X_test)

# imprime a acurácia da base de treino
print('Classificador Árvore de decisão:\n')
print('Acurácia da base de treinamento: {:.2f}'.format(decision_tree.score(X_train, y_train)))

# imprime o classification report da previsão e do teste
print(classification_report(y_test, y_pred, target_names=class_names))

# aplica matrix de confusão na previsão e no teste
conf_matrix = confusion_matrix(y_test, y_pred)

# criar a tabela em datafram da matrix de confusão, carregando os nomes das features
conf_table = pd.DataFrame(data=conf_matrix, index=class_names, columns=[x + "(prev)" for x in class_names])
print(conf_table)
```

## Random Forest

```
# chamando o metodo Random Forest
random_forest = RandomForestClassifier(n_estimators=10)

# treinando a base de treino
random_forest = random_forest.fit(X_train, y_train)

# prevendo a base de test
y_pred = random_forest.predict(X_test)

# imprime a acurácia da base de treino
print('Classificador Random Forest:\n Number of estimators = 10\n')
print('Acurácia da base de treinamento: {:.2f}'.format(random_forest.score(X_train, y_train)))

# imprime o classification report da previsão e do teste
print(classification_report(y_test, y_pred, target_names=class_names))

# aplica matrix de confusão na previsão e no teste
conf_matrix = confusion_matrix(y_test, y_pred)

# criar a tabela em datafram da matrix de confusão, carregando os nomes das features
conf_table = pd.DataFrame(data=conf_matrix, index=class_names, columns=[x + "(prev)" for x in class_names])
print(conf_table)
```

## Extra tree

```
# chamando o metodo Extra tree
extra_tree = ExtraTreesClassifier(n_estimators=10)

# treinando a base de treino
extra_tree = extra_tree.fit(X_train, y_train)

# prevendo a base de test
y_pred = extra_tree.predict(X_test)

# imprime a acurácia da base de treino
print('Classificador Extreme Tree:\n Number of estimators = 10\n')
print('Acurácia da base de treinamento: {:.2f}'.format(extra_tree.score(X_train, y_train)))

# imprime o classification report da previsão e do teste
print(classification_report(y_test, y_pred, target_names=class_names))

# aplica matrix de confusão na previsão e no teste
conf_matrix = confusion_matrix(y_test, y_pred)

# criar a tabela em datafram da matrix de confusão, carregando os nomes das features
conf_table = pd.DataFrame(data=conf_matrix, index=class_names, columns=[x + "(prev)" for x in class_names])
print(conf_table)
```

## Adaboost

```
# chamando o metodo Adaboost
ada_boost = AdaBoostClassifier(tree.DecisionTreeClassifier(max_depth=1),
                               learning_rate=0.1, algorithm="SAMME", n_estimators=5)

# treinando a base de treino
ada_boost.fit(X_train, y_train)

# prevendo a base de test
y_pred = ada_boost.predict(X_test)

# imprime a acurácia da base de treino
print('Classificador AdaBoost:\n Number os estimators = 5\n')
print('Acurácia da base de treinamento: {:.2f}'.format(ada_boost.score(X_train, y_train)))

# imprime o classification report da previsão e do teste
print(classification_report(y_test, y_pred, target_names=class_names))

# aplica matrix de confusão na previsão e no teste
conf_matrix = confusion_matrix(y_test, y_pred)

# criar a tabela em datafram da matrix de confusão, carregando os nomes das features
conf_table = pd.DataFrame(data=conf_matrix, index=class_names, columns=[x + "(prev)" for x in class_names])
print(conf_table)
```

```
# chamando o metodo Adaboost
ada_boost = AdaBoostClassifier(tree.DecisionTreeClassifier(max_depth=1),
                                learning_rate=0.01, n_estimators=5)

# treinando a base de treino
ada_boost.fit(X_train, y_train)

# prevendo a base de test
y_pred = ada_boost.predict(X_test)

# imprime a acurácia da base de treino
print('Classificador AdaBoost:\n Number os estimators = 5\n')
print('Acurácia da base de treinamento: {:.2f}'.format(ada_boost.score(X_train, y_train)))

# imprime o classification report da previsão e do teste
print(classification_report(y_test, y_pred, target_names=class_names))

# aplica matrix de confusão na previsão e no teste
conf_matrix = confusion_matrix(y_test, y_pred)

# criar a tabela em datafram da matrix de confusão, carregando os nomes das features
conf_table = pd.DataFrame(data=conf_matrix, index=class_names, columns=[x + "(prev)" for x in class_names])
print(conf_table)
```

## Redes neurais com uma camada oculta

```
# chamando o metodo Redes neurais
mlp = MLPClassifier(solver='lbfgs', random_state=0, hidden_layer_sizes=[100], learning_rate_init=0.01, max_iter=200000)

# treinando a base de treino
mlp.fit(X_train, y_train)

# prevendo a base de test
y_pred = mlp.predict(X_test)

# imprimindo os parâmetros da rede neural
print('Camadas da rede: {}'.format(mlp.n_layers_))
print('Neurônios na hidden layer: {}'.format(mlp.hidden_layer_sizes))
print('Neurônios na layer de saída: {}'.format(mlp.n_outputs_))
print('Weights na layer de entrada: {}'.format(mlp.coefs_[0].shape))
print('Weights na hidden layer: {}'.format(mlp.coefs_[1].shape))

# imprime a acurácia da base de treino
print('Acurácia da base de treinamento: {:.2f}'.format(mlp.score(X_train, y_train)))

# imprime a acurácia da base de teste
print('Acurácia da base de teste: {:.2f}'.format(mlp.score(X_test, y_test)))

# imprime o classification report da previsão e do teste
print(classification_report(y_test, y_pred, target_names=class_names))

# aplica matrix de confusão na previsão e no teste
conf_matrix = confusion_matrix(y_test, y_pred)

# criar a tabela em datafram da matrix de confusão, carregando os nomes das features
conf_table = pd.DataFrame(data=conf_matrix, index=class_names, columns=[x + "(prev)" for x in class_names])
print(conf_table)
```

## Redes neurais com duas camada oculta

```
# chamando o metodo Redes neurais
mlp = MLPClassifier(solver='lbfgs', random_state=0, hidden_layer_sizes=[100, 60], learning_rate_init=0.01, max_iter=200000)

# treinando a base de treino
mlp.fit(X_train, y_train)

# prevendo a base de test
y_pred = mlp.predict(X_test)

# imprimindo os parâmetros da rede neural
print('Camadas da rede: {}'.format(mlp.n_layers_))
print('Neurônios na camada oculta: {}'.format(mlp.hidden_layer_sizes))
print('Neurônios na camada de saída: {}'.format(mlp.n_outputs_))
print('Pesos na camada de entrada: {}'.format(mlp.coefs_[0].shape))
print('Pesos na camada oculta: {}'.format(mlp.coefs_[1].shape))

# imprime a acurácia da base de treino
print('Acurácia da base de treinamento: {:.2f}'.format(mlp.score(X_train, y_train)))

# imprime a acurácia da base de teste
print('Acurácia da base de teste: {:.2f}'.format(mlp.score(X_test, y_test)))

# imprime o classification report da previsão e do teste
print(classification_report(y_test, y_pred, target_names=class_names))

# aplica matrix de confusão na previsão e no teste
conf_matrix = confusion_matrix(y_test, y_pred)

# criar a tabela em datafram da matrix de confusão, carregando os nomes das features
conf_table = pd.DataFrame(data=conf_matrix, index=class_names, columns=[x + '(prev)' for x in class_names])
print(conf_table)
```
